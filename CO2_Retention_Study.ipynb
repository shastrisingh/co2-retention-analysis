# Carbonation Retention Study    
## Summary:  
- Initially, a visual analysis of CO2 retention over time, as well as CO2 loss over time was performed for each combination of Packsize, Cap Color, CapperHead, MoldNumber and Line present in the data. Subsequently, ANOVA tests were performed to identify from a statistical standpoint whether there were any significant differences in CO2 retention amongst Cap Colors. Throughout, it was found that there were no significant differences at a 5% level of significance.  
- The analysis also yielded that there was no significant difference in CO2 retention by CapperHeads nor MoldNumbers at a 5% level of significance.  
- Thereafter, from the data, it was observed that the 355ml packsize was ran on both lines 3 and 4. This prompted an analysis to ascertain whether there was any differences in CO2 retention based on the lines for the controlled packsize. It was found that there was a statistically significant difference between CO2 retention for 355ml packsize in lines 3 and 4 with line 3 losing approximately 1.85 (46.52%) and line 4 losing approximately 1.61 (45.66%).  
- Additionally, the top 3 CapperHeads/MoldNumbers for each combination of Packsize, Line and Cap Color was also produced to identify the CapperHeads/MoldNumbers that have lowest CO2 loss (highest retention) over the time period.    
  

## Limitations:  
- Filling temperature, Measurement temperature, CO2 Injection Pressure, Flow Rate and Cap Torque variables were not included in the data and hence could not be controlled for.  
- There was also some irregularity with respect to the dates when the readings were taken. That is, not all readings were taken exactly 1 week (7 days) apart.  
- The flavor was not controlled as varying flavors were included.  
- There was no information/data provided on any micro-leaks, CO2 solubility variation due to warming, overfilling/underfilling nor torque slippage.  
Due to these potentially unaccounted factors, there may be confounding, underlying variables attributing to CO2 Loss/Retention that could not be included nor assessed further.

```python
# --- Imports ---
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# --- Helper: Convert week label to days ---
def week_to_days(week):
    if str(week).lower() == "initial":
        return 0
    elif str(week).lower() in ["24hrs", "24 hours"]:
        return 1
    else:
        try:
            return int(week) * 7
        except:
            return None

# --- Core Function: Load, process, analyze, and plot ---
def process_sheet(sheet_name, group_by_col, color_map):
    # Load data
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    df["Week_Days"] = df["Week"].apply(week_to_days)

    # Get all columns with the grouping column prefix
    relevant_cols = [col for col in df.columns if group_by_col in col]
    df["Avg_CO2"] = df[relevant_cols].mean(axis=1)

    # Group and compute CO2 difference
    grouped = (
        df.groupby(["Packsize", "Line", "Cap Color", "Week_Days"], as_index=False)
        .agg({"Avg_CO2": "mean"})
        .sort_values(by=["Packsize", "Line", "Cap Color", "Week_Days"])
    )
    grouped["CO2_Diff"] = grouped.groupby(["Packsize", "Line", "Cap Color"])["Avg_CO2"].diff()
    grouped["CO2_Diff"] = -1 * grouped["CO2_Diff"]
    grouped["CO2_Diff"] = grouped["CO2_Diff"].fillna(0)

    # Define (Packsize, Line) combos to analyze
    combos = [
        ("355ml", 3),
        ("355ml", 4),
        ("500ml", 3),
    ]

    for pack, line in combos:
        df_plot = grouped[(grouped["Packsize"] == pack) & (grouped["Line"] == line)]

        fig, axs = plt.subplots(1, 2, figsize=(14, 6))
        fig.suptitle(f"CO₂ Retention and Differences By {sheet_name} - Busta {pack}, Line {line}",
                     fontsize=16, fontweight='bold')

        # Left: Average CO₂ Retention
        for cap_color in df_plot["Cap Color"].unique():
            df_cc = df_plot[df_plot["Cap Color"] == cap_color]
            axs[0].plot(df_cc["Week_Days"], df_cc["Avg_CO2"], marker='o',
                        label=f"{cap_color} Retention", color=color_map.get(cap_color, "black"))
        axs[0].set_xlabel("Time (Days)")
        axs[0].set_ylabel("Average CO₂ Level")
        axs[0].legend()
        axs[0].grid(True)

        # Right: CO₂ Difference
        for cap_color in df_plot["Cap Color"].unique():
            df_cc = df_plot[df_plot["Cap Color"] == cap_color]
            axs[1].plot(df_cc["Week_Days"], df_cc["CO2_Diff"], marker='o',
                        label=f"{cap_color} Loss", color=color_map.get(cap_color, "black"))
        axs[1].set_xlabel("Time (Days)")
        axs[1].set_ylabel("CO₂ Difference")
        axs[1].legend()
        axs[1].grid(True)

        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

        # Print Percentage CO₂ Loss Summary
        print(f"\n--- Percentage CO₂ loss for {sheet_name} {pack} Line {line} ---")
        for cap_color in df_plot["Cap Color"].unique():
            df_cc = df_plot[df_plot["Cap Color"] == cap_color]
            initial = df_cc[df_cc["Week_Days"] == df_cc["Week_Days"].min()]["Avg_CO2"].mean()
            latest = df_cc[df_cc["Week_Days"] == df_cc["Week_Days"].max()]["Avg_CO2"].mean()
            if initial and latest:
                pct_diff = ((initial - latest) / initial) * 100
                print(f"Cap Color {cap_color}: {pct_diff:.2f}% CO₂ loss from initial to latest week")
            else:
                print(f"Cap Color {cap_color}: Insufficient data to calculate CO₂ loss")
        print(f"\n---------------------------------------------------------------\n")

# --- File path and Color Map ---
file_path = r"C:\Users\shastri.doodnath\OneDrive - S.M. Jaleel & Company Limited\Desktop\CO2 Retention Data Entry Sheet.xlsx"

color_map = {
    "Red": "red",
    "Blue": "blue",
    "Green": "green",
    "Yellow": "gold",
}

# --- Run Analyses ---
process_sheet("CapperHead", "CapperHead", color_map)
process_sheet("MoldNumber", "MoldNumber", color_map)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
import warnings
warnings.filterwarnings("ignore")

# --- Reuse week_to_days from Cell 1 ---
def week_to_days(week):
    if str(week).lower() == "initial":
        return 0
    elif str(week).lower() in ["24hrs", "24 hours"]:
        return 1
    else:
        try:
            return int(week) * 7
        except:
            return None

# --- Load file ---
file_path = r"C:\Users\shastri.doodnath\OneDrive - S.M. Jaleel & Company Limited\Desktop\CO2 Retention Data Entry Sheet.xlsx"

# --- CapperHead Analysis ---
df_capper = pd.read_excel(file_path, sheet_name="CapperHead")
df_capper["Week_Days"] = df_capper["Week"].apply(week_to_days)

capper_cols = [col for col in df_capper.columns if col.startswith("CapperHead")]

df_capper_long = df_capper.melt(id_vars=["Week_Days"], value_vars=capper_cols,
                                var_name="CapperHead", value_name="CO2")

latest_week_capper = df_capper_long["Week_Days"].max()
df_capper_latest = df_capper_long[df_capper_long["Week_Days"] == latest_week_capper].dropna()

groups_capper = [group["CO2"].values for name, group in df_capper_latest.groupby("CapperHead")]
anova_result_capper = f_oneway(*groups_capper)

print("\n--- CapperHead ANOVA ---")
if anova_result_capper.pvalue < 0.05:
    print(f"Significant difference between CapperHeads (p = {anova_result_capper.pvalue:.4f})")
else:
    print(f"No significant difference between CapperHeads (p = {anova_result_capper.pvalue:.4f})")

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_capper_latest, x="CapperHead", y="CO2")
plt.xticks(rotation=90)
plt.title("CO2 Retention by CapperHead (Most Recent Week)")
plt.ylabel("CO2")
plt.xlabel("CapperHead")
plt.tight_layout()
plt.show()

# --- MoldNumber Analysis ---
df_mold = pd.read_excel(file_path, sheet_name="MoldNumber")
df_mold["Week_Days"] = df_mold["Week"].apply(week_to_days)

mold_cols = [col for col in df_mold.columns if col.startswith("MoldNumber")]

df_mold_long = df_mold.melt(id_vars=["Week_Days"], value_vars=mold_cols,
                            var_name="MoldNumber", value_name="CO2")

latest_week_mold = df_mold_long["Week_Days"].max()
df_mold_latest = df_mold_long[df_mold_long["Week_Days"] == latest_week_mold].dropna()

groups_mold = [group["CO2"].values for name, group in df_mold_latest.groupby("MoldNumber")]
anova_result_mold = f_oneway(*groups_mold)

print("\n--- MoldNumber ANOVA ---")
if anova_result_mold.pvalue < 0.05:
    print(f"Significant difference between MoldNumbers (p = {anova_result_mold.pvalue:.4f})")
else:
    print(f"No significant difference between MoldNumbers (p = {anova_result_mold.pvalue:.4f})")

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_mold_latest, x="MoldNumber", y="CO2")
plt.xticks(rotation=90)
plt.title("CO2 Retention by MoldNumber (Most Recent Week)")
plt.ylabel("CO2")
plt.xlabel("MoldNumber")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings("ignore")

def run_anova_tukey(df, sheet_name):
    # Rename to avoid formula issues
    df = df.rename(columns={"Cap Color": "Cap_Color"})
    
    combos = [
        ("355ml", 3),
        ("355ml", 4),
        ("500ml", 3),
    ]
    
    for pack, line in combos:
        subset = df[(df['Packsize'] == pack) & (df['Line'] == line)].dropna(subset=['Avg_CO2', 'Cap_Color'])
        
        print(f"\n\n=== ANOVA for {sheet_name}: Busta {pack}, Line {line} ===")
        
        if subset['Cap_Color'].nunique() > 1:
            model = ols('Avg_CO2 ~ C(Cap_Color)', data=subset).fit()
            anova_table = sm.stats.anova_lm(model, typ=2)
            print(anova_table)
            
            p_val = anova_table['PR(>F)'][0]
            if p_val < 0.05:
                tukey = pairwise_tukeyhsd(endog=subset['Avg_CO2'], groups=subset['Cap_Color'], alpha=0.05)
                print("\nTukey HSD post-hoc test results:")
                print(tukey.summary())
            else:
                print("No significant difference between Cap Colors in this group.")
        else:
            print("Not enough Cap Color groups with sufficient data for ANOVA.")

# --- Load data ---
file_path = r"C:\Users\shastri.doodnath\OneDrive - S.M. Jaleel & Company Limited\Desktop\CO2 Retention Data Entry Sheet.xlsx"

capper_df = pd.read_excel(file_path, sheet_name="CapperHead")
mold_df = pd.read_excel(file_path, sheet_name="MoldNumber")

# Calculate Avg_CO2 for CapperHead
capper_cols = [col for col in capper_df.columns if "CapperHead" in col]
capper_df["Avg_CO2"] = capper_df[capper_cols].mean(axis=1)

# Calculate Avg_CO2 for MoldNumber
mold_cols = [col for col in mold_df.columns if "MoldNumber" in col]
mold_df["Avg_CO2"] = mold_df[mold_cols].mean(axis=1)

# Run analyses
run_anova_tukey(capper_df, "CapperHead")
run_anova_tukey(mold_df, "MoldNumber")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Reuse week_to_days helper
def week_to_days(week):
    if str(week).lower() == "initial":
        return 0
    elif str(week).lower() in ["24hrs", "24 hours"]:
        return 1
    else:
        try:
            return int(week) * 7
        except:
            return None

def run_anova_and_plot(df, sheet_name):
    df = df.rename(columns={"Cap Color": "Cap_Color"})
    
    # Filter for 355ml, Lines 3 and 4
    subset = df[(df['Packsize'] == '355ml') & (df['Line'].isin([3, 4]))].dropna(subset=['Avg_CO2', 'Line', 'Week_Days'])
    
    # Perform ANOVA: Line effect on Avg_CO2 over time (including Week_Days as covariate)
    model = ols('Avg_CO2 ~ C(Line) + Week_Days', data=subset).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    print(f"=== ANOVA: 355ml Line 3 vs Line 4 CO₂ Retention Over Time ({sheet_name}) ===")
    print(anova_table)
    
    p_value = anova_table.loc["C(Line)", "PR(>F)"]
    if p_value < 0.05:
        conclusion = f"There is a statistically significant difference in CO₂ retention over time between Line 3 and Line 4 for 355ml (p = {p_value:.4f})."
    else:
        conclusion = f"There is no statistically significant difference in CO₂ retention over time between Line 3 and Line 4 for 355ml (p = {p_value:.4f})."
    print("\nConclusion:", conclusion)
    
    color_map = {
        "Red": "red",
        "Blue": "blue",
        "Green": "green",
        "Yellow": "gold",
    }
    
    plt.figure(figsize=(10, 6))
    
    for cap_color in subset["Cap_Color"].unique():
        for line in subset["Line"].unique():
            df_line = subset[(subset["Cap_Color"] == cap_color) & (subset["Line"] == line)]
            if not df_line.empty:
                label = f"{cap_color} - Line {line}"
                plt.plot(
                    df_line["Week_Days"],
                    df_line["Avg_CO2"],
                    marker='o',
                    label=label,
                    color=color_map.get(cap_color, "black"),
                    linestyle='-' if line == 3 else '--'
                )
    
    plt.xlabel("Days Since Start")
    plt.ylabel("Average CO₂ Level")
    plt.title(f"CO₂ Retention Over Time for 355ml – Line 3 vs Line 4 ({sheet_name})")
    plt.legend(title="Cap Color and Line")
    plt.grid(True)
    plt.show()

# Load data
file_path = r"C:\Users\shastri.doodnath\OneDrive - S.M. Jaleel & Company Limited\Desktop\CO2 Retention Data Entry Sheet.xlsx"
capper_df = pd.read_excel(file_path, sheet_name="CapperHead")
mold_df = pd.read_excel(file_path, sheet_name="MoldNumber")

# Compute Week_Days
capper_df["Week_Days"] = capper_df["Week"].apply(week_to_days)
mold_df["Week_Days"] = mold_df["Week"].apply(week_to_days)

# Compute average CO₂
capper_cols = [col for col in capper_df.columns if "CapperHead" in col]
capper_df["Avg_CO2"] = capper_df[capper_cols].mean(axis=1)

mold_cols = [col for col in mold_df.columns if "MoldNumber" in col]
mold_df["Avg_CO2"] = mold_df[mold_cols].mean(axis=1)

# Run for CapperHead
run_anova_and_plot(capper_df, "CapperHead")

# Run for MoldNumber
run_anova_and_plot(mold_df, "MoldNumber")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

def week_to_days(week):
    if str(week).lower() == "initial":
        return 0
    elif str(week).lower() in ["24hrs", "24 hours"]:
        return 1
    else:
        try:
            return int(week) * 7
        except:
            return None

# === PREPARE grouped_all FOR CAPPERHEAD if not already done ===
# Load CapperHead sheet and prepare grouped_all dataframe
file_path = r"C:\Users\shastri.doodnath\OneDrive - S.M. Jaleel & Company Limited\Desktop\CO2 Retention Data Entry Sheet.xlsx"

capper_df = pd.read_excel(file_path, sheet_name="CapperHead")
capper_df["Week_Days"] = capper_df["Week"].apply(week_to_days)
capper_cols = [col for col in capper_df.columns if col.startswith("CapperHead")]
capper_df["Avg_CO2"] = capper_df[capper_cols].mean(axis=1)

grouped_all = (
    capper_df
    .groupby(["Packsize", "Line", "Cap Color", "Week_Days"], as_index=False)
    .agg({"Avg_CO2": "mean"})
    .sort_values(by=["Packsize", "Line", "Cap Color", "Week_Days"])
)
grouped_all["CO2_Diff"] = grouped_all.groupby(["Packsize", "Line", "Cap Color"])["Avg_CO2"].diff()
grouped_all["CO2_Diff"] = -1 * grouped_all["CO2_Diff"]
grouped_all["CO2_Diff"] = grouped_all["CO2_Diff"].fillna(0)

# --- CapperHead ANOVA & Tukey for CO2_Diff ---
df_355 = grouped_all[(grouped_all['Packsize'] == '355ml') & (grouped_all['Line'].isin([3,4]))]

model = ols('CO2_Diff ~ C(Line)', data=df_355).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("ANOVA on CO2 Differences (Loss) between Line 3 and Line 4 for 355ml (CapperHead):")
print(anova_table)

p_val = anova_table['PR(>F)'][0]

if p_val < 0.05:
    tukey = pairwise_tukeyhsd(endog=df_355['CO2_Diff'], groups=df_355['Line'], alpha=0.05)
    print("\nTukey HSD results:")
    print(tukey.summary())
    conclusion = f"Significant difference detected between Line 3 and Line 4 in CO₂ loss (p = {p_val:.4f})."
else:
    conclusion = f"No significant difference detected between Line 3 and Line 4 in CO₂ loss (p = {p_val:.4f})."

print("\nConclusion:", conclusion)

avg_loss = (
    df_355
    .groupby(['Week_Days', 'Line'], as_index=False)['CO2_Diff']
    .mean()
)

plt.figure(figsize=(10, 6))
for line in sorted(avg_loss['Line'].unique()):
    df_line = avg_loss[avg_loss['Line'] == line]
    linestyle = '-' if line == 3 else '--'
    plt.plot(df_line['Week_Days'], df_line['CO2_Diff'], marker='o', linestyle=linestyle, label=f'Line {line}')
plt.xlabel('Time (Days)')
plt.ylabel('Average CO₂ Loss')
plt.title('Average CO₂ Loss Over Time for 355ml - Line 3 vs Line 4 (CapperHead)')
plt.legend(title='Production Line')
plt.grid(True)
plt.show()

# --- MoldNumber analysis ---

mold_df = pd.read_excel(file_path, sheet_name="MoldNumber")
mold_df["Week_Days"] = mold_df["Week"].apply(week_to_days)
mold_cols = [col for col in mold_df.columns if col.startswith("MoldNumber")]
mold_df["Avg_CO2"] = mold_df[mold_cols].mean(axis=1)

grouped_mold = (
    mold_df
    .groupby(["Packsize", "Line", "Week_Days"], as_index=False)
    .agg({"Avg_CO2": "mean"})
    .sort_values(by=["Packsize", "Line", "Week_Days"])
)
grouped_mold["CO2_Diff"] = grouped_mold.groupby(["Packsize", "Line"])["Avg_CO2"].diff()
grouped_mold["CO2_Diff"] = -1 * grouped_mold["CO2_Diff"]
grouped_mold["CO2_Diff"] = grouped_mold["CO2_Diff"].fillna(0)

df_355_mold = grouped_mold[(grouped_mold["Packsize"] == "355ml") & (grouped_mold["Line"].isin([3,4]))]

model_mold = ols("CO2_Diff ~ C(Line)", data=df_355_mold).fit()
anova_table_mold = sm.stats.anova_lm(model_mold, typ=2)

print("\nANOVA on CO2 Differences (Loss) between Line 3 and Line 4 for 355ml (MoldNumber):")
print(anova_table_mold)

p_val_mold = anova_table_mold["PR(>F)"][0]

if p_val_mold < 0.05:
    tukey_mold = pairwise_tukeyhsd(endog=df_355_mold["CO2_Diff"], groups=df_355_mold["Line"], alpha=0.05)
    print("\nTukey HSD results (MoldNumber):")
    print(tukey_mold.summary())
    conclusion_mold = f"Significant difference detected between Line 3 and Line 4 in CO₂ loss (p = {p_val_mold:.4f})."
else:
    conclusion_mold = f"No significant difference detected between Line 3 and Line 4 in CO₂ loss (p = {p_val_mold:.4f})."

print("\nConclusion (MoldNumber):", conclusion_mold)

avg_loss_mold = (
    df_355_mold
    .groupby(["Week_Days", "Line"], as_index=False)["CO2_Diff"]
    .mean()
)

plt.figure(figsize=(10, 6))
for line in sorted(avg_loss_mold["Line"].unique()):
    df_line = avg_loss_mold[avg_loss_mold["Line"] == line]
    linestyle = "-" if line == 3 else "--"
    plt.plot(df_line["Week_Days"], df_line["CO2_Diff"], marker="o", linestyle=linestyle, label=f"Line {line}")
plt.xlabel("Time (Days)")
plt.ylabel("Average CO₂ Loss")
plt.title("Average CO₂ Loss Over Time for 355ml - Line 3 vs Line 4 (MoldNumber)")
plt.legend(title="Production Line")
plt.grid(True)
plt.show()
```

##Top 3 CapperHeads per Packsize, Line and Color

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# For CapperHead analysis
df = capper_df.rename(columns={"Cap Color": "Cap_Color"})

# Identify CapperHead columns
capper_cols = [col for col in df.columns if 'CapperHead' in col]

# Melt to long format
melted = df.melt(
    id_vars=['Packsize', 'Line', 'Cap_Color', 'Week_Days'],
    value_vars=capper_cols,
    var_name='CapperHead',
    value_name='CO2_Value'
).dropna(subset=['CO2_Value'])

results = []

for (pack, line, color, capper), group in melted.groupby(['Packsize', 'Line', 'Cap_Color', 'CapperHead']):
    group = group.dropna(subset=['Week_Days', 'CO2_Value'])
    if len(group) >= 2:
        X = group['Week_Days'].values.reshape(-1, 1)
        y = group['CO2_Value'].values
        model = LinearRegression().fit(X, y)
        slope = model.coef_[0]
        results.append({
            'Packsize': pack,
            'Line': line,
            'Cap_Color': color,
            'CapperHead': capper,
            'Slope_CO2_vs_Time': slope
        })

results_df = pd.DataFrame(results)

# Rank capper heads: higher slope better (less loss)
results_df['Rank'] = results_df.groupby(['Packsize', 'Line', 'Cap_Color'])['Slope_CO2_vs_Time']\
    .rank(ascending=False, method='min')

results_df = results_df.sort_values(['Packsize', 'Line', 'Cap_Color', 'Rank'])

top3_per_group = results_df.groupby(['Packsize', 'Line', 'Cap_Color']).apply(
    lambda grp: grp.nsmallest(3, 'Rank')
).reset_index(drop=True)

print("\n" + "="*70)
print("=== TOP 3 CAPPERS BY CO2 SLOPE (higher = better retention) ===")
print("="*70)
for (pack, line), group in top3_per_group.groupby(['Packsize', 'Line']):
    print(f"\nFor Busta {pack} in Line {line}:")
    for color, subgrp in group.groupby('Cap_Color'):
        top_heads = subgrp.sort_values('Rank')['CapperHead'].tolist()
        print(f"\t{color}:\t{', '.join(top_heads)}")
print("="*70 + "\n")


# For MoldNumber analysis
df_mold = mold_df.rename(columns={"Cap Color": "Cap_Color"})

mold_cols = [col for col in df_mold.columns if 'MoldNumber' in col]

melted_mold = df_mold.melt(
    id_vars=['Packsize', 'Line', 'Cap_Color', 'Week_Days'],
    value_vars=mold_cols,
    var_name='MoldNumber',
    value_name='CO2_Value'
).dropna(subset=['CO2_Value'])

results_mold = []

for (pack, line, color, mold), group in melted_mold.groupby(['Packsize', 'Line', 'Cap_Color', 'MoldNumber']):
    group = group.dropna(subset=['Week_Days', 'CO2_Value'])
    if len(group) >= 2:
        X = group['Week_Days'].values.reshape(-1, 1)
        y = group['CO2_Value'].values
        model = LinearRegression().fit(X, y)
        slope = model.coef_[0]
        results_mold.append({
            'Packsize': pack,
            'Line': line,
            'Cap_Color': color,
            'MoldNumber': mold,
            'Slope_CO2_vs_Time': slope
        })

results_mold_df = pd.DataFrame(results_mold)

results_mold_df['Rank'] = results_mold_df.groupby(['Packsize', 'Line', 'Cap_Color'])['Slope_CO2_vs_Time']\
    .rank(ascending=False, method='min')

results_mold_df = results_mold_df.sort_values(['Packsize', 'Line', 'Cap_Color', 'Rank'])

top3_per_group_mold = results_mold_df.groupby(['Packsize', 'Line', 'Cap_Color']).apply(
    lambda grp: grp.nsmallest(3, 'Rank')
).reset_index(drop=True)

print("\n" + "="*70)
print("=== TOP 3 MOLDS BY CO2 SLOPE (higher = better retention) ===")
print("="*70)
for (pack, line), group in top3_per_group_mold.groupby(['Packsize', 'Line']):
    print(f"\nFor Busta {pack} in Line {line}:")
    for color, subgrp in group.groupby('Cap_Color'):
        top_molds = subgrp.sort_values('Rank')['MoldNumber'].tolist()
        print(f"\t{color}:\t{', '.join(top_molds)}")
print("="*70 + "\n")
```
